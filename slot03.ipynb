```python
#Thống kể dữ liệu (Mô tả dữ liệu)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dulieuxettuyendaihoc.csv')
```

```python
print(f'{df["M1"].min()}\n{df["M1"].max()}\n')  # In ra giá trị nhỏ nhất và lớn nhất của cột "M1"
```

```python
print(f'{df["M1"].mean()}\n{df["M1"].median()}\n')  # In ra giá trị trung bình và trung vị của cột "M1"
```

```python
print(f'{df["KT"].mode()}\n')  # In ra giá trị mode (giá trị xuất hiện nhiều nhất) của cột "KT"
```

```python
print(f'{df["M1"].quantile(0.25)}\n')  # In ra giá trị phân vị thứ 25% của cột "M1"
```

```python

df['M1'].max() - df['M1'].min()  # Tính biên độ giá trị (range) của cột "M1"
```

```python

df['M1'].quantile(0.75) - df['M1'].quantile(0.25)  # Tính IQR (khoảng tứ phân vị) của cột "M1"
```

```python

np.mean(np.abs(df['M1'] - np.mean(df['M1'])))  # Tính MAD (độ lệch tuyệt đối trung bình) của cột "M1"
```

```python

df[['M1','M2']].max() - df[['M1','M2']].min()  # Tính biên độ giá trị (range) của các cột "M1" và "M2"
```

```python

df['M1'].var()  # Tính phương sai của cột "M1"
```

```python

df['M1'].std()  # Tính độ lệch chuẩn của cột "M1"
```

```python

print(df[['M1','M2']].mean())  # In ra giá trị trung bình của các cột "M1" và "M2"
```

```python
print(df[['M1','M2']].std())  # In ra độ lệch chuẩn của các cột "M1" và "M2"
```

```python

df[['M1','M2']].std()/df[['M1','M2']].mean()  # Tính hệ số biến thiên (CV) của các cột "M1" và "M2"
```

```python

df['M1'].describe()  # In ra các thống kê tổng quan của cột "M1"
```

```python

df[['M1','M2','M3']].describe()  # In ra các thống kê tổng quan của các cột "M1", "M2" và "M3"
```

```python

cv = df[['M1','M2','M3']].std() / df[['M1','M2','M3']].mean()  # Tính hệ số biến thiên (CV) của các cột "M1", "M2" và "M3"
list(cv)  # Chuyển kết quả hệ số biến thiên thành danh sách
```

```python

df.groupby('GT')['M1'].describe()  # Tính thống kê tổng quan của cột "M1" theo từng nhóm "GT"
```

```python

df.groupby('GT')['M2'].std()/df.groupby('GT')['M2'].mean()  # Tính hệ số biến thiên của cột "M2" theo từng nhóm "GT"
```

```python

df['M1'].hist()  # Vẽ biểu đồ histogram cho cột "M1"
plt.show()
```

```python

df['M1'].hist(bins=14)  # Vẽ histogram của cột "M1" với 14 bin
plt.show()
```

```python

sns.displot(df, x='M1', kind='kde')  # Vẽ biểu đồ mật độ kernel cho cột "M1"
plt.show()
```

```python

sns.displot(data = df[['M1','M2','M3']], kind='kde')  # Vẽ biểu đồ mật độ kernel cho các cột "M1", "M2", "M3"
plt.show()
```

```python

sns.displot(df, x='M1', hue='GT', kind='kde')  # Vẽ biểu đồ mật độ kernel của "M1" phân loại theo "GT"
plt.show()

```

```python

df['M1'].skew()  # Tính độ lệch của phân phối cột "M1"
```

```python

df[['M1','M2','M3']].kurtosis()  # Tính độ nhọn của các cột "M1", "M2", "M3"
```

```python

sns.boxplot(data=df['M1'], orient="h")  # Vẽ boxplot cho cột "M1"
plt.show()
```

```python

sns.boxplot(data=df[['M1','M2','M3']], orient='h')  # Vẽ boxplot cho các cột "M1", "M2", "M3"
plt.show()
```

```python

sns.boxplot(x='M1', y='KT',data=df, orient='h')  # Vẽ boxplot của "M1" so với "KT"
plt.show()
```

```python

sns.boxplot(x='M1', y='GT',data=df, orient='h')  # Vẽ boxplot của "M1" so với "GT"
plt.show()
```

```python

sns.boxplot(x='KT', y='M1',hue='GT',data=df)  # Vẽ boxplot của "M1" so với "KT", phân loại theo "GT"
plt.show()
```

```python

from scipy import stats
stats.probplot(df['M1'],plot=sns.mpl.pyplot)  # Vẽ biểu đồ xác suất chuẩn (QQ plot) cho cột "M1"
plt.show()
```

```python

df[['M1','M2']].cov()  # Tính ma trận hiệp phương sai giữa cột "M1" và "M2"
```

```python

df[['M1','M2']].corr()  # Tính ma trận tương quan giữa cột "M1" và "M2"
```

```python

sns.lmplot(data=df, x='M1', y='M2', fit_reg=True)  # Vẽ biểu đồ hồi quy tuyến tính giữa "M1" và "M2"
plt.show()
```

```python

df[['M1','M2','M3']].corr()  # Tính ma trận tương quan giữa các cột "M1", "M2" và "M3"
```

```python

sns.heatmap(df[['M1','M2','M3']].corr(), vmax=1.0, square=False).xaxis.tick_top()  # Vẽ heatmap cho ma trận tương quan của "M1", "M2", "M3"
```

```python

sns.lmplot(data=df, x='M1', y='M2',hue='GT', fit_reg=True)  # Vẽ biểu đồ hồi quy tuyến tính của "M1" và "M2", phân loại theo "GT"
plt.show()
```

```python

sns.pairplot(df[['M1','M2','M3']],diag_kind='kde', kind='reg')  # Vẽ biểu đồ quan hệ cặp giữa các cột "M1", "M2", "M3"
plt.show()
```
